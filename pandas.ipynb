```python
import pandas as pd
import numpy as np

# Create a DataFrame with random values
np.random.seed(42)  # For reproducibility
df = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])

# Introduce NaN values randomly
df.iloc[2, 1] = np.nan
df.iloc[5, 3] = np.nan
df.iloc[7, 0] = np.nan
df.iloc[9, 2] = np.nan

# Function to highlight NaN values
def highlight_nan(s):
    return ['background-color: yellow' if pd.isna(v) else '' for v in s]

# Apply styling to highlight NaN values
styled_df = df.style.apply(highlight_nan, axis=1)

# Display the styled DataFrame
styled_df

```

```python
import pandas as pd
import numpy as np

# Create a DataFrame with random values
np.random.seed(42)  # For reproducibility
df = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])

# Define a function to highlight specific columns with different colors
def highlight_columns(s):
    color_map = {
        'A': 'background-color: lightblue;', 
        'B': 'background-color: lightgreen;', 
        'C': 'background-color: lightcoral;', 
        'D': 'background-color: lightgoldenrodyellow;'
    }
    return [color_map.get(s.name, '')] * len(s)

# Apply styling
styled_df = df.style.apply(highlight_columns)

# Display the styled DataFrame
styled_df
```
